```python
# 求相关矩阵

# (1)数据标准化
from sklearn import preprocessing

df = preprocessing.scale(df)

df
```

```python
# (2)求相关系数矩阵
covx = np.around(np.corrcoef(df.T), decimals=3)

covx
```

```python
# (3)求解特征值和特征向量
featValue, featVec=  np.linalg.eig(covx.T)  #求解系数相关矩阵的特征值和特征向量

featValue, featVec
```

```python
# 数据不标准化做法
# (1)求均值
def meanX(dataX):
    return np.mean(dataX, axis=0)# axis=0表示依照列来求均值。假设输入list,则axis=1
average = meanX(df)

average
```

```python
# (2)查看列数和行数
m, n = np.shape(df)
m, n
```

```python
# (3)写出同数据集一样的均值矩阵
data_adjust = []
avgs = np.tile(average, (m, 1))
avgs
```

```python
# (4)对数据集进行去中心化
data_adjust = df - avgs

data_adjust
```

```python
# (5)计算协方差阵
covX = np.cov(data_adjust.T)   #计算协方差矩阵

covX
```

```python
# (6)计算协方差阵的特征值和特征向量
featValue, featVec=  np.linalg.eig(covX)  #求解协方差矩阵的特征值和特征向量
featValue, featVec
```

```python
# (7)对特征值进行排序并输出  降序
featValue = sorted(featValue)[::-1]
featValue
```

```python
# (8)绘制散点图和折线图
# 同样的数据绘制散点图和折线图
plt.scatter(range(1, df.shape[1] + 1), featValue)
plt.plot(range(1, df.shape[1] + 1), featValue)
 
# 显示图的标题和xy轴的名字
# 最好使用英文，中文可能乱码
plt.title("Scree Plot")  
plt.xlabel("Factors")
plt.ylabel("Eigenvalue")
 
plt.grid()  # 显示网格
plt.show()  # 显示图形
```

```python
# (9)求特征值的贡献度
gx = featValue/np.sum(featValue)
gx
```

```python
# (10)求特征值的累计贡献度
lg = np.cumsum(gx)
lg
```

```python
# (11)选出主成分
k=[i for i in range(len(lg)) if lg[i]<0.85]
k = list(k)

k
```

```python
# (12)选出主成分对应的特征向量矩阵
selectVec = np.matrix(featVec.T[k]).T
selectVe=selectVec*(-1)
selectVec
```

```python
# (13)求主成分得分
finalData = np.dot(data_adjust,selectVec)
finalData
```

```python
# (14)绘制热力图
# 绘图
 
plt.figure(figsize = (14,14))
ax = sns.heatmap(selectVec, annot=True, cmap="BuPu")
 
# 设置y轴字体大小
ax.yaxis.set_tick_params(labelsize=15)
plt.title("Factor Analysis", fontsize="xx-large")
 
# 设置y轴标签
plt.ylabel("Sepal Width", fontsize="xx-large")
# 显示图片
plt.show()
 
# 保存图片
# plt.savefig("factorAnalysis", dpi=500)
```
